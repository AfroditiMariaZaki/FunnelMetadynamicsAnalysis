### Free Energy heatmap

```python
#Plot the Free Energy Surface as a function of the collective variables p, d for a Funnel Metadynamics simulation#
%matplotlib notebook

path=f"/biggin/b134/bioc1550/Documents/WORK/GlyR/FUNNEL_METAD/PROTEIN_LIGAND/TWOSUBUNITS_LIGAND/RUN{run}"

Z_dat=np.loadtxt(f"{path}/fes.dat")[:, 2]/4.186798
X_dat=np.loadtxt(f"{path}/fes.dat")[:, 0]*10
Y_dat=np.loadtxt(f"{path}/fes.dat")[:, 1]*10

leg=[f"RUN{run}"]

# Convert from pandas dataframes to numpy arrays
X, Y, Z, = np.array([]), np.array([]), np.array([])
for i in range(len(X_dat)):
        X = np.append(X, X_dat[i])
        Y = np.append(Y, Y_dat[i])
        Z = np.append(Z, Z_dat[i])

# create x-y points to be used in heatmap
xi = np.linspace(X.min(), X.max(), 1000)
yi = np.linspace(Y.min(), Y.max(), 1000)

# Z is a matrix of x-y values
zi = griddata((X, Y), Z, (xi[None,:], yi[:,None]), method='cubic')

# I control the range of my colorbar by removing data 
# outside of my range of interest
zmin = 0
zmax = int(Z_dat[0])
zi[(zi<zmin) | (zi>zmax)] = None
zstep=2
levels = np.arange(zmin, zmax, zstep)
fig, ax1=plt.subplots()

# Create the contour plot
cbar = plt.contourf(xi, yi, zi, 15, cmap=plt.cm.rainbow,
                  levels=levels)

xmin=int(np.amin(X_dat))
xmax=int(np.amax(X_dat))
ymin=int(np.amin(Y_dat))
ymax=int(np.amax(Y_dat))
print(xmin, xmax, ymin, ymax)
leg=[f"RUN{run}"]
ax1.set_xlim([xmin-1.0, xmax+1.0])
ax1.set_ylim([ymin-1, ymax+1])
startx, endx = ax1.get_xlim()
starty, endy = ax1.get_ylim()
ax1.xaxis.set_ticks(np.arange(startx, endx+1, 5))
ax1.yaxis.set_ticks(np.arange(starty, endy+1, 2))
ax1.set_xlabel('Position along funnel axis [$\AA$]', fontsize=large_font)
ax1.set_ylabel('Distance from funnel axis [$\AA$]', fontsize=large_font)
cbar.ax.get_yaxis().labelpad = 20
cbar.ax.tick_params(labelsize=small_font)
#cbar.set_label('Free Energy [kcal/mol]', rotation=270, fontsize=12)
plt.tick_params(axis='both', bottom=True, top=True, right=True, which='major', labelbottom=True, labeltop=False, direction='in', labelsize=small_font)
#ax1.legend(leg, loc=1, fancybox=True, shadow=True)
#plt.text(17.0, 19, f"RUN{run}", fontsize=12, fontstyle='italic', bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round'))
plt.colorbar()  
ax1.set_title("Free Energy [kcal/mol]")
plt.tight_layout() # otherwise the right y-label is slightly clipped
plt.savefig(f"fes{run}-heatmap.png", bbox_inches='tight', dpi=600)
plt.show()
```

### Calculate and plot loop C opening

```python
from MDAnalysis.analysis.distances import dist
u=mda.Universe("conf.gro", "traj_comp.xtc")

distance=[]
time=[]
loopC=u.select_atoms(f"resid {max_res} and name CA and bynum 0-3406")
bsheet=u.select_atoms("resid 40 and name CA and bynum 3407-6813")

# distance = dist(loopC, bsheet, offset=0, box=None).run()
# print(distance.dist)

for ts in u.trajectory:
    distance.append((mda.analysis.distances.dist(loopC, bsheet, offset=0, box=None)))
    time.append((ts.time/1000))

time=np.array(time)
dist=np.array(distance)[:, 2]
print(time, dist)

xmin=float("{0:.0f}".format(np.amin(time)))              
xmax=float("{0:.0f}".format(np.amax(time)))
ymin=float("{0:.2f}".format(np.amin(dist)))
ymax=float("{0:.2f}".format(np.amax(dist)))            
leg=[f"Run{run}"]
# print(f"x-axis minimum and maximum is {xmin} and {xmax}")
# print(f"y-axis minimum and maximum is {ymin} and {ymax}")
              
              
fig, (ax1) = plt.subplots(1,figsize=(8,2))

ax1.plot(time, dist, color='grey', linewidth=1.0)
ax1.set_xlim([xmin, xmax])
ax1.set_ylim([int(ymin)-1, int(ymax)+1])
ax1.xaxis.set_ticks(np.arange(int(xmin), int(xmax), 50))
ax1.yaxis.set_ticks(np.arange(int(ymin)-2, int(ymax)+1, 5))
ax1.tick_params(axis='both', bottom=True, top=True, right=True, which='major', labelbottom=True, labeltop=False, direction='in', labelsize=small_font)
ax1.set_xlabel('Time [ns]', fontsize=large_font)
ax1.set_ylabel('Distance [$\AA$]', fontsize=large_font)
#ax1.legend(leg, loc=2,  fancybox=True, shadow=True)
ax1.set_title=("Loop C Opening")
fig.tight_layout()

plt.savefig(f"{path}/loopC{run}.png", bbox_inches='tight', dpi=600)
```
